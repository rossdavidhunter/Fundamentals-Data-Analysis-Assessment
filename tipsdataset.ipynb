# Fundamentals of Data Analysis Assignment

This jupyter notebook is to complete assisgnment on well -know tips dataset using seaborn python package and jupyter lab notebook.

Data visulaization is very important in data analysis as it helps identify relationships between the variables. For this assisgment we will use the seaborn package and tips dataset. Seaborn is a data visualization libary that creates nice visualtions in the form of plots. To aide this visualization we will also use matplotlib, numpy and pandas packages.

From diagram below Exploratory Data Analysis (EDA) is an approach in analyzing data sets to summarize main charateristics using various visual methods as seen below. Its primary use is to tell us what data means beyond formal modelling or testing tasks. Some of the example techniques used are as follows and will be used in this assignment along with others to provide descriptive statistics:

1. Histograms
2. Scatter Plots
3. Box Plots
4. Violin Plots
5. Pair Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Loading of Dataset from Seaborn

```python
tips = sns.load_dataset("tips")
```

# Shows first 20 lines from dataset

```python
tips.head(20)
```

# Breakdown of Count, Mean, Std Dev etc

```python
tips.describe()
```

```python
tips.shape
```

# Relationship between total bill and tip

```python
sns.relplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, color="purple")
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", kind="line", data=tips)
plt.show()
```

# Histogram

```python
sns.distplot(tips['tip'], kde=True, bins=10)
plt.show()
```

# Scatterplot

```python
sns.catplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
plt.show()
```

# Distplot

```python
sns.distplot(tips['tip'],hist=True, bins=10)
plt.show()
```

# Boxplot

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
plt.show()
```

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot("day", "total_bill", "sex", data=tips, kind="box")
    g.set_axis_labels("Day", "Total Bill")
plt.show()
```

# Violin Plots

```python
sns.catplot(x="day", y="total_bill", hue="sex",kind="violin", data=tips)
plt.show()
```

# Point Plots

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="point", data=tips)
plt.show()
```

# Pairplot

```python
sns.pairplot(tips, hue = "sex", palette="Set1")
plt.show()
```

# Joint Plots

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, color="purple")
plt.show()
```

```python
sns.jointplot(x = tips["total_bill"], y = tips["tip"],kind = "kde", 
color="purple")
plt.show()
```

# Barplots

```python
sns.barplot(x ="sex" , y ="total_bill" , data=tips)
plt.show()
```

```python
sns.barplot(x = "smoker", y = "total_bill", data =tips)
plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", data =tips)
plt.show()
```

# LM Plots

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, hue="day")
plt.show()
```
