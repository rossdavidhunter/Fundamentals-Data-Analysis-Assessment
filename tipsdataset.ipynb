```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading of Dataset from Seaborn

```python
tips = sns.load_dataset("tips")
```

# Shows first 20 lines from dataset

```python
tips.head(20)
```

# Breakdown of Count, Mean, Std Dev etc

```python
tips.describe()
```

```python
tips.shape
```

# Histogram

```python
sns.distplot(tips['tip'], kde=False, bins=10)
plt.show()
```

# Scatterplot

```python
sns.catplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
plt.show()
```

```python
sns.distplot(tips['tip'],hist=False, bins=10)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="sex",kind="violin", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="point", data=tips)
plt.show()
```
