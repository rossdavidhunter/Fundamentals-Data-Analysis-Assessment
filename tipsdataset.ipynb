Some of the example techniques used are as follows and will be used in this assignment along with others to provide descriptive statistics:

1. Histograms
2. Scatter Plots
3. Box Plots
4. Violin Plots
5. Pair Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Loading of Dataset from Seaborn

This will load in tips dataset from seaborn package.

```python
tips = sns.load_dataset("tips")
```

Use of head and tail function will show data at start and end of dataset to give you a feel of what is contained in dataset.

```python
tips.tail(20)
```

# Breakdown of Count, Mean, Std Dev etc

describe() function is used to view some basic statistical details like percentile, mean, std etc. of a data frame or a series of numeric values.

```python
tips.describe()
```

# Shape of dataset

This will return info on number of rows and columns.

```python
tips.shape
```

# Relationship between total bill and tip

```python
sns.relplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, color="purple")
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", kind="line", data=tips)
plt.show()
```

What we can see from first 2 plots above is that the relationship between total bill and tip is showing a strong concentration that when the bill is between 10 and 20, tips between 2 and 4 are given.

The 3rd plot we can see that distribution is too random for tips with respect to total_bill. We can can conclude that the tip is not dependent on the value of total_bill.


# Histogram

Use of histogram plot shows the tips given are concentrated around 2-4 values. Changing of bin values will change the way the data is spread out as seen below.

```python
sns.distplot(tips['tip'], kde=True, bins=10)
plt.show()
sns.distplot(tips['tip'], kde=True, bins=20)
plt.show()
```

# Scatterplot

Use of scatterplot is useful when you want to understand relationship between 2 varaiables. As seen below there is a high concentration of total bill between 10 and 30 for Thurs, Sat and Sun. Friday shows a low concentration around 10-20 mark.

```python
sns.catplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
plt.show()
```

# Boxplot

A Box Plot is a visual representation of five point summary statistics of a given data set. A five number summary includes:
1. Minimum
2. First Quartile
3. Median (Second Quartile)
4. Third Quartile
5. Maximum

You can add in hue to split boxplot for example smokers and non smokers. From below you can see the highest bills are concentrated around smokers.



```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
plt.show()
```

```python
sns.boxplot(x = "day", y = "total_bill", data=tips, hue = "smoker")
plt.show()
```

# Violin Plots

These are similar to boxplots. See image below for comparison.

![](Image.PNG)

```python
sns.catplot(x="day", y="total_bill", hue="sex",kind="violin", data=tips)
plt.show()
```

# Point Plots

A point represents an estimate of central tendancy for a numeric variable. They show only mean values and error rates surrounding those mean values. They are useful in finding change in variable. 
Below you can see that the total bill rises from Thursday to Sunday for males while it decreases from Thursday to Friday, rises on Sat and remains flat across Sunday for females.

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="point", data=tips)
plt.show()
```

# Pairplot

The diagonal part shows the distplot or histogram with kernel density estimation. The upper and lower part of the Pairplot shows the scatterplot. The hue colours the plot using categorical columns.
hue = sex  It will color the plot gender wise.
palette = Set1 is a type of color combination used to color the plot

```python
sns.pairplot(tips, hue = "sex", palette="Set1")
plt.show()
```

# JointPlots

A jointplot takes two variables and as seen below creates a histogram and scatterplot. The second plot is a contour plot and a Kerner Density Estimate (KDE) plot.

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, color="purple")
plt.show()
```

```python
sns.jointplot(x = tips["total_bill"], y = tips["tip"],kind = "kde", 
color="purple")
plt.show()
```

# Barplots

Barplots are meant for plotting categorical columns vs numerical columns. It creates bar in visualisation.

```python
sns.barplot(x ="sex" , y ="total_bill" , data=tips)
plt.show()
```

```python
sns.barplot(x = "smoker", y = "total_bill", data =tips)
plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", data =tips)
plt.show()
```

# LM Plots

sns.lmplot is a plot that fits the regression line to the dataset showing as scatterplots. It follows the ordinary least square method and the line represents the best fit line. 
This shows the linear regression fit of total_bill variable for the different days as shown in the plot legend. This is obtained using hue = day in sns.lmplot.

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, hue="day")
plt.show()
```
